```python
import sys
import os
sys.path.insert(0, os.path.abspath('..'))
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
import funcs
import imp
import random
import mne
imp.reload(funcs)
```

```python
downsample = 600
exp_info = funcs.load_info()
exp_info["alpha pre"] = None
exp_info["alpha post"] = None

load_path = os.path.join('..', '..', '..', 'data', 'subjects_rest_eeg.pkl')
subjects_rest_eeg = pd.read_pickle(load_path)
```

```python
task_time_name = ['pre', 'post']
for subject_id in range(1, 27):
    for tasktime in range (1, 3):
        raw = subjects_rest_eeg[subjects_rest_eeg['subject id'] == subject_id][task_time_name[tasktime-1]].values[0]
        print("\nsubject_id: ", subject_id, " tasktime: ", task_time_name[tasktime-1], "\n")

        raw_downsampled = raw.copy().resample(downsample)
        eeg = raw_downsampled.get_data()[1:32,:]
        alpha = np.zeros((eeg.shape[0]))
        for channel in range(eeg.shape[0]):
            alpha[channel] = funcs.dfa_alpha(eeg[channel,:])

        idx = exp_info[(exp_info['subject id'] == subject_id)].index[0]
        exp_info.at[idx, 'alpha '+task_time_name[tasktime-1]] = alpha
```

```python
save_path = os.path.join('..', '..', '..', 'data', 'subjects_rest_alpha.pkl')
exp_info.to_pickle(save_path)
```
